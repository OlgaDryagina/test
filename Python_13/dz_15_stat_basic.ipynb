# Задание 1. Базовое изучение
Изучить представленный набор данных на основе описания его столбцов и выбрать 8 столбцов для дальнейшего изучения (среди них должны быть как числовые, так и категориальные). Провести расчет базовых метрик для них, кратко описать результаты.

```python
import pandas as pd
```

```python
df_horses = pd.read_csv('horse_data.csv', names = ['surgery', 'age','HospitalNumber', 'RectalTemperature', 'pulse', 'RespiratoryRate','TemperatureOfExtremities',
             'peripheralPulse',' mucousMembranes', 'capillaryRefillTime', 'pain', 'peristalsis', 'abdominalDistension',
             'nasogastricTube','nasogastricReflux', 'nasogastricRefluxPH','rectalExamination','abdomen','packedCellVolume',
              'totalProtein','abdominocentesisAppearance','abdomcentesisTotalProtein','outcome', 'surgicalLesion', 'typeOfLesion',
              'typeOfLesion_1', 'typeOfLesion_2','cpData'], na_values='?')
df_horses.head()

#параметр na_values='?' заменяет ? на Nan
```

```python
df_horses_part = df_horses[['surgery','age', 'RectalTemperature','pulse', 'RespiratoryRate','abdominalDistension', 'totalProtein', 'outcome']]
df_horses_part.head()
```

Определим, как распределяются значения столбца Операция:

```python
surgery_counts = {}

for surgery in df_horses_part.surgery.round():
    if surgery not in surgery_counts:
        surgery_counts[surgery] = 1
    else:
        surgery_counts[surgery] += 1
        
surgery_counts    
```

Данные по столбцу Операция интерпретируются следующим образом: 1 = Да, была операция, 2 = Лечение прошло без хирургического вмешательства. Наблюдаем пропуск в данных.        

Определим, как распределяются значения столбца Возраст.

```python
age_counts = {}

for age in df_horses_part.age.round():
    if age not in age_counts:
        age_counts[age] = 1
    else:
        age_counts[age] += 1
        
age_counts        
        
```

Согласно описания Базы данных по конским коликам: 1 = взрослая лошадь, 2 = молодая (< 6 месяцев). По полученным результатам 1 встречается 276 раз, 9 - 24 раза. Будем считать, что 9 это опечатка в данных (вместо цифры 9 должны были проставит 2).


Определим, как распределяются значения столбца Вздутие живота:

```python
abdominalDistension_counts = {}

for abdominalDistension in df_horses_part.abdominalDistension.round():
    if abdominalDistension not in abdominalDistension_counts:
        abdominalDistension_counts[abdominalDistension] = 1
    else:
        abdominalDistension_counts[abdominalDistension] += 1
        
abdominalDistension_counts    
```

Данные столбца Вздутие живота интерпретируются следующим образом: 1 = нет, 2 = легкий, 3 = умеренный, 4 = тяжелый. В данных по этому столбцу наблюдаем много пропусков (из 300 заполнено 244).

# Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

Найдем выбросы в столбце RectalTemperature:

```python
print(df_horses_part.RectalTemperature.mean())
print(df_horses_part[(df_horses_part.RectalTemperature > 35) & (df_horses_part.RectalTemperature < 42)].RectalTemperature.mean())
```

По столбцу RectalTemperature выбросов не наблюдается, данные в пределах реально допустимых показателей температуры тела.

Найдем выбросы по столбцу pulse через межквартильный размах:

```python
q1 = df_horses_part.pulse.quantile(0.25)
q3 = df_horses_part.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_horses_part[df_horses_part.pulse.between(lower_bound, upper_bound, inclusive=True)]  #inclusive=True это включительно
remove_outliers
```

Посмотрим строки, которые содержат данные, которые мы принимаем за выбросы


```python
pd.concat([df_horses_part, remove_outliers]).drop_duplicates(keep=False)
```

Таким образом, нашли строки, в которых пульс от 150 до 184 ударов при норме 30-40. Эти данные оставляем, так как мы исследуем данные о состоянии больных животных, которые могут имет повышенную частоту сердечных сокращений.


Найдем выбросы по столбцу RespiratoryRate

```python
q1 = df_horses_part.RespiratoryRate.quantile(0.25)
q3 = df_horses_part.RespiratoryRate.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_ = df_horses_part[df_horses_part.RespiratoryRate.between(lower_bound, upper_bound, inclusive=True)]  #inclusive=True это включительно
remove_outliers_
```

```python
pd.concat([df_horses_part, remove_outliers_]).drop_duplicates(keep=False).head()
```

Нормальная частота дыхания взрослой лошади 7-10 раз. Больныеживотные могут иметь учащенное дыхание, поэтому все, что определили как выбросы, оставляем в данных.


Найдем выбросы по столбцу totalProtein.

```python
q1 = df_horses_part.totalProtein.quantile(0.25)
q3 = df_horses_part.totalProtein.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_1 = df_horses_part[df_horses_part.totalProtein.between(lower_bound, upper_bound, inclusive=True)]  #inclusive=True это включительно
remove_outliers_1
```

```python
pd.concat([df_horses_part, remove_outliers_1]).drop_duplicates(keep=False).head()
```

Нормальный показатель белка 6.5-7.0. У больных животных он может быть повышен, поэтому оставляем в данных.
